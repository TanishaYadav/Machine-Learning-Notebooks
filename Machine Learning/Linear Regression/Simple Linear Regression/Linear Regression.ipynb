```python
X = dataset.YearsExperience
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

#  Gradient Descent

###### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_new = sc_X.fit_transform(np.array(X_train).reshape(-1,1))   # requires 2-D array as input
y_new = sc_X.transform(np.array(y_train).reshape(-1,1))
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01
    m_arr = np.array([m_curr])
    b_arr = np.array([b_curr])
    cost_arr = np.array([10])
    print(len(x), len(y))
    
    
    for i in range(iterations):
        y_pred = (m_curr * x) + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_pred)]) 
        dm = -(2/n) * sum(2 * x * (y - y_pred))
        db = -(2/n) * sum(y - y_pred)
        m_curr = m_curr - learning_rate * dm
        b_curr = b_curr - learning_rate * db
#         m_arr = np.append(m_arr,m_curr)
#         b_arr = np.append(b_arr,b_curr)
#         cost_arr = np.append(cost_arr,cost)
#     plt.scatter(m_arr, cost_arr)
#     plt.show()
#     plt.scatter(b_arr,cost_arr, color = 'r')
#     plt.show()
#     return (m_arr[-1],b_arr[-1])
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))
    
    
```

```python
gradient_descent(X_new, y_new)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(np.array(X_train).reshape(-1,1), y_train)

              # fit method takes X as an 2-D array so we reshaped it, (-1 means unknown dimension)
    #since series don't have reshape method we converted it into numpy array
```

```python
y_pred = regressor.predict(np.array(X_test).reshape(-1,1))
```

```python
plt.scatter(X_train, y_train, color = 'r')
plt.plot(X_train , regressor.predict(np.array(X_train).reshape(-1,1)), color ='b')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Experience')
plt.ylabel('Salary')


```

```python
plt.scatter(X_test, y_test, color = 'r')
plt.plot(X_test , y_pred, color ='b')           # regression line remains same even if we used X_train
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
regressor.score(np.array(X_test).reshape(-1,1),y_test)
```
